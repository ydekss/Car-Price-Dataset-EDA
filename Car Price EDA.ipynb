```python
# Plotting a bar graph for the values
colors = ['pink', 'silver']
explode = (0.05, 0.05)
df['doornumber'].value_counts().plot(kind ='pie', y='Number of Clusters', autopct='%1.0f%%', colors=colors, explode=explode)
```

#### There are more number of four wheeler cars in the data

```python
# Plotting a Bar Graph for the Count of Sentiment
df['fueltype'].value_counts().plot(kind = 'bar', color ='purple', width = 0.4)
plt.xlabel('fueltype')
plt.ylabel('Count')
plt.title('Count of fueltype')
plt.legend()
```

#### Gas cars comapratively have a huge market compared to diesel cars 

```python
sns.set_theme(style="whitegrid")
sns.relplot(data=df, x="price", y="horsepower", hue="fueltype")
```

#### We can clearly see a positive relationship between price and horsepower

```python
sns.set_theme(style="whitegrid")
sns.relplot(data=df, x="horsepower", y="citympg", hue="fueltype",)
```

#### We can see a negative relationship between Horsepower and City Mielage

```python
df.groupby('aspiration')['doornumber'].value_counts().plot(kind='bar', color = 'pink')
```

#### The most common cars have a Std aspirator and have 4 doors

```python
df.groupby('aspiration').mean()['compressionratio'].plot(kind = 'bar', color ='pink', width = 0.4)
plt.xlabel('Aspiration')
plt.ylabel('Compression')
```

#### Turbo cars generate more compression compared to std cars

```python
df.groupby('carbody')['citympg'].mean().plot(kind='bar',color ='red', width = 0.4)
plt.xlabel('Carbody')
plt.ylabel('City Milage')
plt.show()
df.groupby('carbody')['highwaympg'].mean().plot(kind='bar',color ='pink', width = 0.4)
plt.xlabel('Carbody')
plt.ylabel('Highway Milage')
plt.show()
```

#### By comparing these 2 graphs we can conclude that Cars are giving more mileage on Highway as compared to City

```python
colors = ['pink', 'silver']
explode = (0.05, 0.05,0.05)
df.groupby('drivewheel')['price'].max().plot(kind ='pie', y='Number of Clusters', autopct='%1.0f%%', colors=colors, explode=explode)
```

#### RWD drive have the highest price compared to the other drivewheel cars

```python
df.groupby('drivewheel')['cylindernumber'].value_counts().plot(kind='bar', color = 'black')
```

#### The most common car in the dataframe is a foward drivewheel car with 4 cylinders

```python
plt.boxplot(df['horsepower'])
```

#### There are only 2 outliers with over 200 Horsepower and most of the cars are having the HP in the range of 70 - 120
